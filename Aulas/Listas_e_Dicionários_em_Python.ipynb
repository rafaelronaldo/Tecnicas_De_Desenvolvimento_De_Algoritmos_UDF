# O que são Listas?
*   Uma coleção ordenada e mutável de itens
*   Os itens podem ser de tipos diferentes (strings, números, booleanos, etc.)
*   São definidas por colchetes []
*   Cada item tem uma posição chamada de índice, começando em 0





```python
frutas = ["maçã", "banana", "laranja"]
numeros = [1, 2, 3, 4, 5]
misturado =["texto, 10, True"]
```

# Acessando e Modificando Listas

## Acessando Elementos

```python
frutas = ["maçã", "banana", "laranja"]
#acessar primeiro elemento
print(frutas[0])
```

```python
#acessar o último elemento
print(frutas[-1])
```

```python
#acessar o intervalo(slice)
print(frutas[0:2])
```

## **Modificando Elementos**

```python
#alterar um elemento
frutas[0] = "morango"
print(frutas)
```

```python
#adicionar um elemento
frutas.append("abacaxi")
print(frutas)
```

```python
#remover um elemento
frutas.remove("banana")
print(frutas)
```

# O que são Dicionários?



*   Uma coleção não ordenada de pares chave-valor
*   As chaves devem ser únicas e imutáveis (strings, números ou tuplas)
*   São definidas por chaves {}
*   Permitem acessar valores de forma rápida e eficiente usando suas chaves





## Acessando e Modificando Dicionários

## Acessando Valores


```python
contato = {
    "nome": "João",
    "idade": 30,
    "cidade": "São Paulo"
}
#acessa valor pela chave
print(contato["nome"])
```

```python
#acessar outra chave
print(contato["idade"])
```

```python
#usar get() para acesso seguro
print(contato.get("profissão", "N/A"))
```

## Modificando Dicionário


```python
#alterar um valor
contato["idade"] = 31
print(contato["idade"])
```

```python
#adicionar novo par chave-valor
contato["profissao"] = "Desenvolvedor"
print(contato)
```

```python
#remover um par chave-valor
del contato["cidade"]
print(contato)
```

# Métodos úteis de Dicionários

```python
#retorna todas as chaves
contato = {
"nome": "João",
"idade": 30,
}
print(contato.keys())
```

```python
#retorna todos os valores
contato = {
"nome": "João",
"idade": 30,
}
print(contato.values())
```

```python
#retorna pares chave-valor
contato = {
"nome": "João",
"idade": 30,
}
print(contato.items())
```

```python
#método get() = Acesso Seguro
contato={"nome": "João", "idade": 30}
print(contato.get("nome"))
```
