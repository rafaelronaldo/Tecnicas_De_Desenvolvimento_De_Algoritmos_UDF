## Gerenciamento de Coordenadas (Lista de Tuplas)

```python
#==================================================================
# MINI PROJETO 2: Gerencimento de Coordenadas (Lista de Tuplas)
#Objetivo: Armazenar sequências de valores fixos (latitude, longitude, nome)
#==================================================================

#Lista (locais): Coleção de todas as coordenadas registradas.
#Tupla: Cada tupla (imútavel) armazena uma coordenada e seu nome.

locais = [
   # Tupla 1: (latitude, longitude, nome_do_local)
   (-23.55052, -46.63330, "São Paulo"),
   #Tupla 2
   (-15.7801, -47.9292, "Brasília"),
   #Tupla 3
   (40.7128, -74.0060, "Nova York")
]

print("\n2. Gerencimento de Coordenadas Geográficas")
print("-" * 40)

# 1. Adicionar Nova Coordenada
print("-> Adicionando nova coordenada (Tupla)...")
# Cria uma nova tupla para garantir que a coordenada seja registrada de forma imutável.
novo_local = (34.0522, -118.2437, "Los Angeles")
locais.append(novo_local)  # A lista é mutável, permitindo adicionar a nova tupla.
print(f"Local {novo_local[2]} adicionado.")

# 2. Acessar e Desempacotar (Unpack) uma Tupla
print("\n-> Desempacotando a última coordenada...")
# Desempacota os elementos da última tupla da lista em variáveis separadas.
lat, lon, nome = locais[-1]
print(f"Latitude: {lat}, Longitude: {lon}, Local: {nome}")

# 3. Iterar e Formatar Coordenadas
print("\n-> Coordenadas registradas:")
for coordenada in locais:
    # Acessa os elementos da tupla usando índices.
    print(f"Local: {coordenada[2]} | Lat/Lon: ({coordenada[0]}, {coordenada[1]})")

# Importante: Tuplas são imutáveis; tentar alterar um elemento resultaria em um erro.
print("-" * 40)

# Aplicações: Mapeamento e geolocalização (onde as coordenadas não devem mudar), atributos de classes, sistemas de cache.
```

```python
# Passo 1: Definir equipes, potes e confederações
# Representaremos os potes como listas de nomes de equipes (strings).
# Em um cenário real, você provavelmente buscaria os rankings da FIFA para determinar os potes.
# Para esta simulação, usaremos equipes e confederações de exemplo.

# Dicionário para mapear equipes às suas confederações
equipes_confederacoes = {
    "Brazil": "CONMEBOL", "Argentina": "CONMEBOL", "France": "UEFA", "Germany": "UEFA",
    "Spain": "UEFA", "Belgium": "UEFA", "Portugal": "UEFA", "England": "UEFA",
    "Netherlands": "UEFA", "Croatia": "UEFA", "Italy": "UEFA", "Mexico": "CONCACAF",
    "USA": "CONCACAF", "Colombia": "CONMEBOL", "Uruguay": "CONMEBOL", "Switzerland": "UEFA",
    "Senegal": "CAF", "Japan": "AFC", "Iran": "AFC", "Serbia": "UEFA",
    "Poland": "UEFA", "South Korea": "AFC", "Morocco": "CAF", "Tunisia": "CAF",
    "Canada": "CONCACAF", "Australia": "AFC", "Ecuador": "CONMEBOL", "Ghana": "CAF",
    "Cameroon": "CAF", "Saudi Arabia": "AFC", "Qatar": "AFC", "Wales": "UEFA"
}


# Pot 1 (Top seeds)
pot1 = ["Brazil", "Argentina", "France", "Germany", "Spain", "Belgium", "Portugal", "England"]

# Pot 2
pot2 = ["Netherlands", "Croatia", "Italy", "Mexico", "USA", "Colombia", "Uruguay", "Switzerland"]

# Pot 3
pot3 = ["Senegal", "Japan", "Iran", "Serbia", "Poland", "South Korea", "Morocco", "Tunisia"]

# Pot 4
pot4 = ["Canada", "Australia", "Ecuador", "Ghana", "Cameroon", "Saudi Arabia", "Qatar", "Wales"]

print("Potes definidos:")
print("Pot 1:", pot1)
print("Pot 2:", pot2)
print("Pot 3:", pot3)
print("Pot 4:", pot4)

# Passo 2: Simular o sorteio
import random

# Definir o número de grupos (assumindo 8 grupos para uma simulação mais simples)
num_grupos = 8
grupos = {f"Grupo {i+1}": [] for i in range(num_grupos)}

# Lista de potes
potes = [pot1, pot2, pot3, pot4]

# Sortear as equipes de cada pote
for i, pot in enumerate(potes):
    random.shuffle(pot) # Embaralha as equipes no pote
    for equipe in pot:
        # Encontrar um grupo válido para a equipe
        grupo_encontrado = False
        while not grupo_encontrado:
            grupo_sorteado = random.choice(list(grupos.keys())) # Sorteia um grupo
            # Verificar se a equipe pode ser adicionada a este grupo (regra da confederação)
            pode_adicionar = True
            confederacao_equipe = equipes_confederacoes[equipe]
            for equipe_no_grupo in grupos[grupo_sorteado]:
                if equipes_confederacoes[equipe_no_grupo] == confederacao_equipe:
                    pode_adicionar = False
                    break # Não pode adicionar se já houver equipe da mesma confederação

            # Regra especial para a UEFA: até duas equipes da UEFA por grupo
            if confederacao_equipe == "UEFA":
                count_uefa = sum(1 for team in grupos[grupo_sorteado] if equipes_confederacoes[team] == "UEFA")
                if count_uefa >= 2:
                    pode_adicionar = False


            if pode_adicionar:
                grupos[grupo_sorteado].append(equipe)
                grupo_encontrado = True

print("\nSorteio realizado. Grupos provisórios:")
for grupo, equipes in grupos.items():
    print(f"{grupo}: {equipes}")
```

# Catálogo de Produtos com Busca Rápida (Dicionário)

```python
# MINI PROJETO 3: Catálogo de Produtos com Busca Rápida (Dicionário)
# Objetivo: Mapear um ID de produto (chave) para os detalhes do produto (valor).

# Dicionário (catalogo): Estrutura principal. [cite: 146]
# Chave: ID do produto (string). Valor: Dicionário contendo detalhes.

catalogo = {
    "P001": {"nome": "Teclado Mecânico", "estoque": 25, "preco": 350.00},
    "P002": {"nome": "Monitor LED 27", "estoque": 18, "preco": 1200.00},
    "P003": {"nome": "Mouse Gamer", "estoque": 40, "preco": 150.00}
}

print("\n3. Catálogo de Produtos com Busca Rápida")
print("-" * 40)

# 1. Adicionar Novo Produto (CREATE)
novo_produto_id = "P004"
catalogo[novo_produto_id] = {"nome": "Webcam HD", "estoque": 50, "preco": 250.00}  # Adiciona um novo par chave:valor.
print(f"Produto {catalogo[novo_produto_id]['nome']} adicionado ao catálogo.")

# 2. Buscar por Chave (Busca O(1) Rápida) (READ)
id_busca = "P002"
# Acessar diretamente pelo ID (chave) é a forma mais rápida de busca em um dicionário.
if id_busca in catalogo:
    produto_detalhes = catalogo[id_busca]
    print(f"\nDetalhes do Produto {id_busca}:")
    print(f"Nome: {produto_detalhes['nome']}, Preço: R$ {produto_detalhes['preco']:.2f}")
else:
    print(f"\nProduto com ID {id_busca} não encontrado.")

# 3. Atualizar Estoque (UPDATE)
id_update = "P001"
catalogo[id_update]["estoque"] = 5  # Acessa o dicionário interno pela chave e atualiza o valor.
print(f"\nEstoque do {catalogo[id_update]['nome']} atualizado. Novo estoque: {catalogo[id_update]['estoque']}")
print("-" * 40)
```

# Agrupamento de Vendas por Região (Dicionários Aninhados e Tuplas)

```python
# MINI PROJETO 4: Agrupamento de Vendas por Região (Dicionários Aninhados e Tuplas)
# Objetivo: Agrupar dados de vendas (Lista de Tuplas) por uma chave de agrupamento (Região).
# Dicionário (dados_vendas): Chave é a região.
# Valor: Dicionário contendo detalhes anuais e uma lista de registros de vendas.
# Lista de Tuplas: Cada tupla (imutável) é um registro de venda (id_venda, valor).

dados_vendas = {
    "SUDESTE": {
        "total_anual": 150000.00,
        "vendas_detalhe": [
            ("V001", 5000.00),
            ("V002", 12000.00),
            ("V003", 8000.00)
        ]
    },
    "SUL": {
        "total_anual": 95000.00,
        "vendas_detalhe": [
            ("V004", 5500.00),
            ("V005", 10000.00)
        ]
    }
}

print("\n4. Agrupamento de Vendas por Região")
print("-" * 40)

# 1. Adicionar Nova Venda à Região SUDESTE [cite: 240]
regiao = "SUDESTE"
nova_venda = ("V006", 9500.00)

# Acessa o dicionário SUDESTE, depois a chave 'vendas_detalhe' (que é uma lista), e adiciona a nova tupla.
dados_vendas[regiao]["vendas_detalhe"].append(nova_venda)
print(f"Nova venda {nova_venda[0]} adicionada na região {regiao}.")

# 2. Calcular o Valor Total de Vendas Detalhadas no SUDESTE
vendas_sudeste = dados_vendas["SUDESTE"]["vendas_detalhe"]
# Usa compreensão de lista e a função sum() para somar o segundo elemento (valor) de cada tupla.
soma_detalhes = sum(venda[1] for venda in vendas_sudeste)

print(f"\nSoma dos detalhes de vendas SUDESTE: R$ {soma_detalhes:.2f}")
print(f"Total anual registrado: R$ {dados_vendas[regiao]['total_anual']:.2f}")

# 3. Listar as Regiões (Chaves)
print("\nRegiões em análise:")
# O método .keys() retorna as chaves do dicionário principal.
for reg in dados_vendas.keys():
    print(f"- {reg}")
print("-" * 40)

# Aplicações: Relatórios financeiros/BI (estruturar dados hierarquicamente), dentro de classes, API de back-end.
```

# Mapeamento de Níveis de Acesso e Permissões (Dicionário com Tuplas)



```python
# MINI PROJETO 5: Mapeamento de Níveis de Acesso (Dicionário com Tuplas)
# Objetivo: Definir conjuntos de permissões fixas por nível de usuário.

# Dicionário (permissoes): Chave é o nível de acesso.
# Tupla: Valor é uma tupla imutável, contendo as permissões (strings).
permissoes = {
    "ADMIN": ("CREATE_USER", "READ_ALL", "UPDATE_ANY", "DELETE_ALL"),  # Tupla de permissões para Admin
    "MANAGER": ("READ_ALL", "UPDATE_OWN"),  # Tupla de permissões para Manager
    "VIEWER": ("READ_ALL",)  # Tupla de um elemento para Viewer (vírgula obrigatória)
}

print("\n5. Mapeamento de Níveis de Acesso e Permissões")
print("-" * 40)

# 1. Acessar Permissões de um Nível Específico (READ)
nivel_acesso = "ADMIN"

# Acessa a tupla de permissões usando a chave do dicionário.
perms_admin = permissoes.get(nivel_acesso)

if perms_admin:
    print(f"Permissões do nível {nivel_acesso}:")
    # Itera sobre a tupla de permissões.
    for perm in perms_admin:
        print(f"- {perm}")
else:
    print(f"Nível de acesso '{nivel_acesso}' não definido.")

# 2. Verificar uma Permissão Específica
perm_verificar = "DELETE_ALL"
# Usa o operador 'in' para checar se a permissão existe dentro da tupla.
if perm_verificar in permissoes["ADMIN"]:
    print(f"\nADMIN possui a permissão: {perm_verificar}")
else:
    print(f"\nADMIN NÃO possui a permissão: {perm_verificar}")

# 3. Tentativa de Adicionar uma Permissão (Imutabilidade da Tupla)
# print("Tentando adicionar permissão ao VIEWER...")
# permissoes["VIEWER"].append("DELETE_OWN") # Erro: 'tuple' object has no attribute 'append'
# A tupla garante a integridade do conjunto de permissões.
print("\nA Integridade do conjunto de permissões é garantida pela Tupla (Imutabilidade).")
print("-" * 40)

# Aplicações: Sistemas de autenticação e autorização (RBAC), atributos fixos em classes, arquivos de configuração.
```
