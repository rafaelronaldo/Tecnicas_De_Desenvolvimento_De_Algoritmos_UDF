# Se√ß√£o 2: E-commerce

## 2.1 Cat√°logo de Produtos (Lista de Tuplas)

```python
# Cat√°logo usa uma Lista de Tuplas (imut√°veis)
catalogo = [
    ("Smartphone", 1500, "Eletr√¥nicos"),
    ("T√™nis", 250, "Moda"),
    ("Notebook", 3500, "Eletr√¥nicos"),
    ("Livro", 40, "Educa√ß√£o"),
    ("Fone de ouvido", 200, "Eletr√¥nicos")
]

# Filtrar produtos da categoria 'Eletr√¥nicos' usando Compreens√£o de Lista.
eletronicos = [produto for produto in catalogo if produto[2] == "Eletr√¥nicos"]
assert len(eletronicos) == 3
print("Produtos eletr√¥nicos:", eletronicos)
```

## 2.2 Carrinho de Compras (Dicion√°rio)

```python
# Carrinho usa um Dicion√°rio: {produto: quantidade}
carrinho = {}
carrinho["Notebook"] = 1
carrinho["Fone de ouvido"] = 2
del carrinho["Notebook"] # Remove um item do carrinho
assert "Notebook" not in carrinho
assert carrinho["Fone de ouvido"] == 2
print("Carrinho:", carrinho)
```

## 2.3 C√°lculo de Total com Desconto (Listas e Compreens√£o)

```python
precos = [1500, 200, 3500]
# Aplica 10% de desconto (multiplica por 0.9) em todos os pre√ßos
precos_com_desconto = [p * 0.9 for p in precos]
# Calcula o total
total = sum(precos_com_desconto)
assert round(total, 2) == round(1500 * 0.9 + 200 * 0.9 + 3500 * 0.9, 2)
print(f"Total com desconto: R$ {total:.2f}")
```

## 2.4 Estoque por Categoria (Dicion√°rio de Listas)

```python
# Estoque usa Dicion√°rio: {categoria: [produtos]}
estoque = {
    "Eletr√¥nicos": ["Smartphone", "Notebook", "Fone de ouvido"],
    "Moda": ["T√™nis", "Camisa"],
    "Educa√ß√£o": ["Livro"]
}
# Encontra a categoria com mais produtos (maior length da lista)
categoria_maior = max(estoque, key=lambda c: len(estoque[c]))
assert categoria_maior == "Eletr√¥nicos"
print("Categoria com mais produtos:", categoria_maior)
```

# Se√ß√£o 3: Jogos

## 3.1 Ranking de Jogadores (Lista de Tuplas)

```python
# Ranking usa Lista de Tuplas (nome, pontua√ß√£o)
ranking = [("Ana", 120), ("Bruno", 200), ("Carlos", 180)]
# Ordena a lista pela pontua√ß√£o (√≠ndice 1) em ordem decrescente (reverse=True)
ranking.sort(key=lambda x: x[1], reverse=True)
assert ranking[0][0] == "Bruno"
print("Top 3 jogadores:", ranking[:3])
```

## 3.2 Invent√°rio de Itens (Dicion√°rio de Tuplas)

```python
# Invent√°rio usa Dicion√°rio: {item: (quantidade, raridade)}
inventario = {
    "Espada": (1, "raro"),
    "Po√ß√£o": (5, "comum"),
    "Armadura": (2, "√©pico")
}
# Filtra itens com raridade "raro" ou "√©pico"
raros = [item for item, (_, raridade) in inventario.items() if raridade in ["raro", "√©pico"]]
assert "Espada" in raros and "Armadura" in raros
print("Itens raros:", raros)
```

# 3.3 Hist√≥rico de Partidas (Lista de Tuplas e Dicion√°rio)

```python
# Hist√≥rico de partidas: (Jogador 1, Jogador 2, Vencedor)
partidas = [
    ("Ana", "Bruno", "Ana"),
    ("Carlos", "Ana", "Carlos"),
    ("Bruno", "Carlos", "Carlos")
]

# Conta as vit√≥rias usando um Dicion√°rio
vitorias = {}
for partida in partidas:
    vencedor = partida[2]
    # Incrementa a contagem. .get(vencedor, 0) garante que come√ßa em 0 se a chave n√£o existir.
    vitorias[vencedor] = vitorias.get(vencedor, 0) + 1
assert vitorias["Carlos"] == 2
print("Vit√≥rias:", vitorias)
```

## 3.4 Configura√ß√µes do Jogo (Dicion√°rio)

```python
# Configura√ß√µes do jogo usam Dicion√°rio: {config: valor}
config = {
    "volume": 70,
    "dificuldade": "m√©dio",
    "idioma": "portugu√™s"
}

# Altera configura√ß√µes
config["volume"] = 50
config["idioma"] = "ingles"
assert config["volume"] == 50
print("Configura√ß√µes:", config)
```

# Se√ß√£o 4: Playlists

## 4.1 Playlist de M√∫sicas (Lista de Dicion√°rios)

```python
# Playlist usa Lista de Dicion√°rios
playlist = [
    {"nome": "M√∫sica A", "artista": "Artista 1", "duracao": 2.5},
    {"nome": "M√∫sica B", "artista": "Artista 2", "duracao": 3.2},
    {"nome": "M√∫sica C", "artista": "Artista 1", "duracao": 4.0}
]

# Filtra m√∫sicas com dura√ß√£o superior a 3 minutos
longas = [m["nome"] for m in playlist if m["duracao"] > 3]
assert "M√∫sica B" in longas and "M√∫sica C" in longas
print("M√∫sicas com mais de 3 minutos:", longas)
```

## 4.2 Favoritar M√∫sicas (Dicion√°rio)

```python
# Favoritas usa Dicion√°rio: {m√∫sica: status_booleano}
favoritas = {
    "M√∫sica A": False,
    "M√∫sica B": True,
    "M√∫sica C": True
}
# Lista apenas as m√∫sicas onde o valor √© True
favoritas_listadas = [nome for nome, fav in favoritas.items() if fav]
assert len(favoritas_listadas) == 2
print("Favoritas:", favoritas_listadas)
```

## 4.3 Estat√≠sticas da Playlist (Lista de Tuplas)

```python
# M√∫sicas usa Lista de Tuplas (nome, dura√ß√£o)
musicas = [("M√∫sica A", 2.5), ("M√∫sica B", 3.2), ("M√∫sica C", 4.0)]

# Soma a dura√ß√£o (√≠ndice 1 de cada tupla)
total = sum([duracao for nome, duracao in musicas])
# Calcula a m√©dia
media = total / len(musicas)
assert round(media, 2) == round(total / 3, 2)
print(f"Total: {total:.2f} min | M√©dia: {media:.2f} min")
```

## 4.4 Organizar por Artista (Dicion√°rio de Listas)

```python
# Agrupamento usa Dicion√°rio: {artista: [m√∫sicas]}
por_artista = {
    "Artista 1": ["M√∫sica A", "M√∫sica C"],
    "Artista 2": ["M√∫sica B"]
}
# Encontra o artista com o maior n√∫mero de m√∫sicas
artista_top = max(por_artista, key=lambda a: len(por_artista[a]))
assert artista_top == "Artista 1"
```
