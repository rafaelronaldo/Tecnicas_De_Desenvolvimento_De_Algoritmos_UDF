# LISTAS E TUPLAS
São sequências de tipos de dados que armazenam uma coleção ordenada de itens.

**Listas**:

*   mutáveis;
*   ideias para dados dinâmicos;
*   homogêneos.

**Tuplas**:


*   são imutáveis
*   ideais para dados fixos (que não serão alterados)
*   heterogêneos


# Conceitos - Listas
É uma estrutra de dados nativa, versátil e amplamente utilizada em Python.

Permite armazenamento e manipulação de uma sequência ordenada e mutável de
elementos

```python
colors = ["red", "green", "blue"]
colors[0]
```

Listas podem conter tipos de dados heterogêneos (string, int, float, object) e suportam
uma grande quantidade de operações, fazendo com que se tornem essenciais para
organizar e processar dados.

```python
lista = ["nome", 32, 1.75]
print(lista)
```

## Quando usar listas:
Armazenar e organizar sequências de dados;

Percorrer elementos;

Adicionar e remover elementos dinamicamente;

Acessar elementos pelo índice;

Ordenar e inverter coleções.

```python
#retornam um objeto do tipo lista.
lista = list(range(0,10,2))
type (lista)
len(lista)
print(lista)
```

## Métodos de listas

.append() - adiciona um elemento ao fim da lista

.extend() - aumenta a lista acrescentando elementos de um iterável

.insert() - Insere um elemento em uma posição específica

.remove() - Remove a primeira ocorrência do elemento

.pop() - Remove e retorna o elemento de uma posição específica

.clear() - Limpa a lista removendo todos os seus elementos

.index() - Retorna o índice da primeira ocorrência do elemento

.count() - Retorna a quantidade de ocorrências do elemento

.sort() - Ordena a lista

.reverse() - Inverte a lista

.copy() - Cria uma cópia rasa da lista

# ***TUPLAS***

Estrutura de dados nativa do Python que entrega uma sequência imutavel de valores.
Pode guardar qualquer tipo de dado.
São úteis para armazenar dados heterogênos, como registros de banco de dados ou uma coleção fixa de itens.

```python
coordinates = (10.5, 20.3)
person = ("Alice", 30, "Enginner")
```

É possivel criar truplas: Vazias; Com apenas um item; com múltiplos itens.

Quando criada com apenas um item, a tupla deve ser encerrada com,(vírgula).

Apesar de ser possível criá-las sem parêntesis, convém usa-los a fim de facilitar a legibilidade do código.

Quando usar listas: Armazenar coleções fixas de dados correlacionados; Retornar múltiplos valores de uma só função; Agrupar dados que não devem mudar com o passar do tempo.

## Métodos de tuplas

# TAREFA 1
Escreva um programa que solicita que o
usuário digite 5 números inteiros e os coloca
em uma lista.

Quando terminar, imprima a lista.

```python
quantidade_numero = []
for contador in range (5):
  numero = int(input("Digite o {}º número: ".format(contador+1)))
  quantidade_numero.append(numero)
print(quantidade_numero)
```

# TAREFA 2

Escreva um programa que aplica a função
módulo a todos os elementos e uma lista.


```python
lista = [-5, 3, -10, 8, -2]
lista_modulo = []

for numero in lista:
  lista_modulo.append(abs(numero))
print("Lista original: {}".format(lista))
print("Lista em módulo: {}".format(lista_modulo))
```

# TAREFA 3
Escreva um programa que apaga todos os
elementos negativos de uma lista.

```python
lista = [-5, 3, -10, 8, -2]
lista_sem_negativos = []

for contador in lista:
  if contador >= 0:
    lista_sem_negativos.append(contador)
print(lista_sem_negativos)
```

# TAREFA 5
Faça um programa que solicita uma
quantidade indefinida de números positivos
(deve parar quando um número negativo for
digitado).

Armazene-os em duas listas:
uma para os números pares;
outra para os números ímpares.

A seguir, mostre a porcentagem de pares e
ímpares digitados.
Exercícios

```python

lista_par = []
lista_impar = []
numero_positivo = float(input("Digite um número positivo: "))
while numero_positivo >= 0:
  if numero_positivo % 2 == 0:
    lista_par.append(numero_positivo)
  else:
    lista_impar.append(numero_positivo)
print(lista_par)
print(lista_impar)
```

# TAREFA 6
Faça um programa que solicita uma
quantidade indefinida de números positivos
(deve parar quando um número negativo for
digitado).

Armazene todos os números digitados em
uma lista, sem repetição.


```python
lista_positivo = []
numero = int(input("Digite um número: "))
while numero >= 0:
  if numero not in lista_positivo:
    lista_positivo.append(numero)
  numero = int(input("Digite um número: "))
print(lista_positivo)
```
