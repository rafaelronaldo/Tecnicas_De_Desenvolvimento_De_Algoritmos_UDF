# Exercício 1
*   Faça um programa que leia um número n e imprima n linhas na tela
com o seguinte formato (exemplo se n = 5):




```python
numero = int(input("Digite um número:"))
linhas = []
for contador in range(1, numero + 1):
  linhas.append(contador)
  print(linhas)
```

# Exercício 2

*   Faça um programa que leia n notas, mostre as notas e a média.



```python
#Mostra as n notas
notas = []
numero = int(input("Digite a quantidade de notas: "))
for contador in range(numero):
  dados = float(input("Digite a {}ª nota: ".format(contador+1)))
  notas.append(dados)
print(notas)

#Calcula a média
soma = 0
for contador in notas:
  soma = soma + contador
media = soma/numero
print("A média é: {}".format(media))
```

# Exercício 3

*   Lê duas listas com 5 inteiros cada;
*   Checa quais elementos da segunda lista são iguais a algum elemento
da primeira lista;
*   Se não houver elementos em comum, o programa deve informar isso.






```python
lista1 = []
lista2 = []
for contador in range(5):
  lista1.append(int(input()))
print()
for contador in range(5):
  lista2.append(int(input()))
print()

um_elemento_comum = False

#Supomos que não temos elementos comuns
for a in lista1:
  for b in lista2:
    if (a == b):
        um_elemento_comum = True # há elemento comum
        print(str(a))
if not um_elemento_comum:
  print("Não tem elemento comum.")
```
